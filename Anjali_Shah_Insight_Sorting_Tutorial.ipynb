## Selection Sort

```python
# Selectionsort that sorts in increasing order of value of list elements

def selectionsort(input_list):
    
    print 'Unsorted List: ', input_list
    
    # Outer loop - Loop through the list to bring the smallest valued element in ith position
    # The last list element will automatically be in the right position at the end of the sorting algorithm.
    # Hence, outer loops runs only until the element in second to last position
    for i in range(0, (len(input_list) - 1)):
        imin = i
        
        # Inner loop - Loop through the list to find the smallest valued element
        for j in range(i+1, len(input_list)):
            if input_list[j] < input_list[imin]:
                imin = j
        
        # Swap the element in the ith position to newly found smallest element
        input_list[i], input_list[imin] = input_list[imin],input_list[i]
        
        # Elements upto current value of i are in their respective sorted positions
        print 'Intermedial Step %d: ' % (i+1), input_list
    
    # The final sorted list
    print 'Sorted List: ', input_list
```

```python
# Call selectionsort with a list of integers to be sorted in increasing order of value
# Measure running time of selectionsort that has O(n^2) time complexity in worst and average cases

start_time = time()

selectionsort([7,5,9,1,3,2])

end_time = time()

print('Total selectionsort running time: {}'.format(end_time - start_time))
```

```python
# Selectionsort that sorts in increasing order of value of list elements (no print statements)

def selectionsortnoprints(input_list):
     
    # Outer loop - Loop through the list to bring the smallest valued element in ith position
    # The last (highest) list element will automatically be in the right position at the end of the sorting algorithm.
    # Hence, outer loops runs only until the element in second to last position
    for i in range(0, (len(input_list) - 1)):
        imin = i
        
        # Inner loop - Loop through the list to find the smallest valued element
        for j in range(i+1, len(input_list)):
            if input_list[j] < input_list[imin]:
                imin = j
        
        # Swap the element in the ith position to newly found smallest element
        input_list[i], input_list[imin] = input_list[imin],input_list[i]
            
    # The final sorted list
    return input_list

# Plot time complexity for selectionsort by increasing number of elements to be sorted
def plottimecomplexityplots(num_lists):
    
    # List initializations
    ss_input_list = [[] for i in range(num_lists)]
    ss_output_list = [[] for i in range(num_lists)]
    global ss_running_time 
    ss_running_time = [[] for i in range(num_lists)]
    
    # Generate input_lists with random integers, sort the lists and save runtime complexity
    for i in range(num_lists):
        random.seed(10)
        num_ints = ((i+1)*50)
        
        ss_input_list[i] = np.random.choice(num_ints, num_ints, replace=False)
                
        start_time = time()
        ss_output_list[i] = selectionsortnoprints(ss_input_list[i])
        end_time = time()
        
        ss_running_time[i] = end_time - start_time
        
    for i in range(num_lists):
        print 'Elements: %d, ' % len(ss_input_list[i]), 
        print 'Running Time: %f' % ss_running_time[i]
    
    # Plot graphs of runtime complexity by number of elements to be sorted
    plt.plot([len(ss_input_list[i]) for i in range(num_lists)], ss_running_time)
    plt.xlabel('Number of Elements to be Sorted')
    plt.ylabel('Selection Sort Time Complexity - O(n^2)')
    plt.show()
    
# Call the plot function        
plottimecomplexityplots(5)      
```

## Bubble Sort

```python
# Bubblesort that sorts in increasing order of value of list elements

def bubblesort(input_list):
    
    print 'Unsorted List: ', input_list
        
    exchanges = True
    passnum = len(input_list)-1
    step = 1
    # Outer loop - Loop through the list (n - 1) times where n is the list's length
    # The highest list element will automatically be in the right position at the end of each pass.
    # With help of exchanges flag only run the outer loop as long as swaps are necessary
    while passnum > 0 and exchanges:
       exchanges = False
       
       # Inner loop - Loop through the list to bubble up highest element towards end of the list
       for i in range(passnum):
           if input_list[i]>input_list[i+1]:
               exchanges = True
               # Swap the elements in the ith position and (i+1)th position if element in ith position is higher
               input_list[i], input_list[i+1] = input_list[i+1], input_list[i]
       
       # Elements upto current value of step starting at end of list are in their respective sorted positions
       print 'Intermedial Step %d: ' % (step), input_list
       passnum = passnum-1
       step = step+1
                
    # The final sorted list
    print 'Sorted List: ', input_list
```

```python
# Call bubblesort with a list of integers to be sorted in increasing order of value
# Measure running time of bubblesort that has O(n^2) time complexity in worst and average cases

start_time = time()

bubblesort([7,5,9,1,3,2])

end_time = time()

print('Total bubblesort running time: {}'.format(end_time - start_time))
```

## Insertion Sort

```python
# Insertionsort that sorts in increasing order of value of list elements

def insertionsort(input_list):
    
    print 'Unsorted List: ', input_list
        
    # Outer loop - Loop through the list (n - 2) times where n is the list's length
    # The smallest list element will automatically be in its correct position to the left of index at the end of each pass.
    for index in range(1,len(input_list)):

        currentvalue = input_list[index]
        position = index

        # Inner loop - Loop through the list and move elements 
        # Move elements in lower positions and greater than current element to the right
        while position>0 and input_list[position-1]>currentvalue:
            input_list[position]=input_list[position-1]
            position = position-1

        input_list[position]=currentvalue
       
        # Elements upto current value of index are in their respective sorted positions
        print 'Intermedial Step %d: ' % (index), input_list
                       
    # The final sorted list
    print 'Sorted List: ', input_list
```

## Merge Sort

```python
# Mergesort that sorts in increasing order of value of list elements

def mergesort(input_list):
    print("Splitting ",input_list)
    # Only proceed if list has atleast 2 elements, otherwise return original list
    if len(input_list)>1:
        
        # Split list into 2 
        mid = len(input_list)//2
        lefthalf = input_list[:mid]
        righthalf = input_list[mid:]

        # Recursively call mergesort on lefthalf
        mergesort(lefthalf)
        # Recursively call mergesort on righthalf
        mergesort(righthalf)
        # Merge such that all elements are in order of increasing value
        merge(lefthalf, righthalf, input_list)
    else:
        # Print original list if number of elements is less than 2
        print 'Sorted List: ', input_list
    
def merge(lefthalf, righthalf, sorted_list):
    print("Merging ",lefthalf, righthalf)
    # Pointer initializations
    i=0
    j=0
    k=0
    # Look for elements in lefthalf and righthalf such that sorted_list has them in correct order
    while i < len(lefthalf) and j < len(righthalf):
        if lefthalf[i] <= righthalf[j]:
            sorted_list[k]=lefthalf[i]
            i=i+1
        else:
            sorted_list[k]=righthalf[j]
            j=j+1
        k=k+1
    
    # Add leftovers (if any) from lefthalf
    while i < len(lefthalf):
        sorted_list[k]=lefthalf[i]
        i=i+1
        k=k+1
        
    # Add leftovers (if any) from righthalf
    while j < len(righthalf):
        sorted_list[k]=righthalf[j]
        j=j+1
        k=k+1
    
    print 'Sorted List: ', sorted_list
```

## Heap Sort

```python
# Heapsort that sorts in increasing order of value of list elements
def heapsort(input_list):
   
   print 'Unsorted List: ', input_list
   # Number of elements
   num_elems = len(input_list)
   
   # Place input_list in max-heap order
   heapify(input_list, num_elems)
 
   end = num_elems - 1
   while end > 0:
      # Swap the root(maximum value) of the heap with the last element of the heap
      input_list[end], input_list[0] = input_list[0], input_list[end]
        
      print("Max-heapifying: ",input_list)
      
      # Decrement the size of the heap so that the previous max value will stay in its proper place
      end = end - 1
      # Put the heap back in max-heap order
      bubbledown(input_list, 0, end)
   
   # The final sorted list
   print 'Sorted List: ', input_list
        
def heapify(input_list, num_elems):
   
   # Start is assigned the index in input_list of the last parent node
   start = (num_elems - 2) / 2
   
   while start >= 0:
      # Bubble down the node at start index to the proper place such that 
      # all nodes below the start index are in heap order
      bubbledown(input_list, start, num_elems-1)
      start = start - 1
      # after sifting down the root all nodes/elements are in heap order
 
def bubbledown(input_list, start, end):
   
   # end represents the limit of how far down the heap to sift
   root = start

   while root * 2 + 1 <= end:                         # While the root has at least one child
      child = root * 2 + 1                            # root*2+1 points to the left child
      
      # If the child has a sibling and the child's value is less than its sibling's
      if ((child + 1 <= end) and (input_list[child] < input_list[child + 1])): 
         child = child + 1                            # point to the right child instead
      if input_list[root] < input_list[child]:        # out of max-heap order
         input_list[root], input_list[child] = input_list[child], input_list[root]
         root = child                                 # repeat to continue bubbling down the child now
      else:
         return
```

## Radix Sort

```python
# Radixsort that sorts in increasing order of value of list elements

def radixsort(input_list,num_buckets, maxLen):
    
    print 'Unsorted List: ', input_list
    
    # Start with the Least Significant Digit (LSD) and sort every list element by it's LSD
    # Repeat the process for other digits moving from right-to-left 
    # Repeat all the way up to and including Most Significant Digit (MSD)
    for i in range(maxLen):
        buckets = [[] for j in range(num_buckets)]
        for elem in input_list:
            buckets[(elem/10**i)%num_buckets].append(elem)
        print 'Intermediate Step %d: ' % (i+1), buckets
        
        input_list=[]
        for section in buckets:
            input_list.extend(section)
    print 'Sorted List: ', input_list
```

```python
# Call radixsort with a list of integers to be sorted in increasing order of value
# Measure running time of radixsort that has O(nk) time complexity in worst and average cases
# k is the length of the highest(longest) number(string)

start_time = time()

radixsort([792, 539, 911, 145, 312, 289], 10, 3)

end_time = time()

print('Total radixsort running time: {}'.format(end_time - start_time))
```
